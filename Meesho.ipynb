### Cost Price for Delivered & Cancelled:

```python
cp_d=data1['Cost Price'].sum()
cp_d
```

```python
fsa_d=data1['Final Settlement Amount'].sum()
fsa_d
```

```python
fsa_r=data2['Final Settlement Amount'].sum()
fsa_r
```

### GST Credit:

```python
gst=data1['GST on Shipping Charge'].sum()
gst
```

```python
profit_month=((fsa_d-cp_d)-(fsa_r)-(gst))
profit_month
```

```python
sum_cost_price_by_group = similar_sku_data.groupby('SKU')['Cost Price'].sum()
sum_cost_price_by_group
```

```python
sum_final_settlement_by_group = similar_sku_data.groupby('SKU')['Final Settlement Amount'].sum()
sum_final_settlement_by_group
```

```python
profit=sum_final_settlement_by_group-sum_cost_price_by_group
profit
```
